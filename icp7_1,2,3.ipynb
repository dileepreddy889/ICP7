```
import requests
link = 'https://en.wikipedia.org/wiki/Google'
response = requests.get(link)
from bs4 import BeautifulSoup
content = BeautifulSoup(response.content, "html.parser")
text = content.get_text()
#print(text)
f = open('input.txt', 'w',encoding='utf-8')
f.write(text)
```

```
import nltk
```

```
nltk.download('punkt')
```

```
wtokens=nltk.word_tokenize(text)
```

```
for w in wtokens:
  print(w)
```

```
stokens=nltk.sent_tokenize(text)
```

```
for s in stokens:
  print(s)
```

```
nltk.download('averaged_perceptron_tagger') #pos
pos= nltk.pos_tag(stokens)
print(pos)
```

```
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer #stemming
```

```
pStemmer= PorterStemmer()
for w in wtokens:
  print(pStemmer.stem(w))
```

```
from nltk.stem import WordNetLemmatizer
```

```
nltk.download('wordnet')
```

```
lemmatizer = WordNetLemmatizer()
for w in wtokens:
  print(lemmatizer.lemmatize(w))
```

```
from nltk import wordpunct_tokenize, pos_tag, ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```
print(ne_chunk(pos_tag(wordpunct_tokenize(text))))
```
